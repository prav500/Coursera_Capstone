```python
data1 = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',skiprows=0)[0]
```

Convert data frame to List for cleanup

```python
data1_list = data1.values.tolist()
#print (data1_list)
#print(data1_list[1].extend(data1_list[0]))
data1_list_all = data1_list[0] + data1_list[1] + data1_list[2] + data1_list[3] + data1_list[4] +data1_list[5] +data1_list[6] +data1_list[7]+ data1_list[8]

#data1_list_all.remove('Not assigned')
for s in data1_list_all:
    if s.endswith('Not assigned'):
       data_list_all=data1_list_all.remove(s)
       print(s)
```

```python
data1_list_all
```

```python
from pandas import DataFrame
df = DataFrame(data1_list_all,columns=["All_Columns"])
df
#new_dataframe = df[df.endswith('Not assigned')]
#new_dataframe
```

remove not assigned

```python
df = df[df.All_Columns.str.contains("Not assigned") != 1]
df

```

```python
df.shape
```

Split the data in three columns

```python
sub ='('
start = 3
df["Indexes"]= df["All_Columns"].str.find(sub,start) 
df['Postal Code'] = df['All_Columns'].str.slice(0,3)
df['Borough'] = df.apply(lambda df:df['All_Columns'][3:df['Indexes']], axis=1)
df['Neighborhood'] = df.apply(lambda df:df['All_Columns'][df['Indexes']+1:-1], axis=1)
```

```python
df
```

replace '/' in Neighborhood by ','

```python
#df['Borough'] = df.apply(lambda df:df['All_Columns'][3:df['Indexes']], axis=1)
df['Neighborhood'] = df.apply(lambda df:df['Neighborhood'].replace('/',','),axis=1)
df
```

```python
df=df.drop(5)
df.head()
```

```python
df=df.drop(['All_Columns','Indexes'], axis=1)
```

required cleaned up dataframe is ready

```python
df
```

download coordinated file in a dataframe

```python
df_lat_lng = pd.read_csv('http://cocl.us/Geospatial_data/Geospatial_Coordinates.csv')
df_lat_lng


```

merge two dataframe to obtain Latitude and Longitude on main dataframe 

```python
result = pd.merge(df,df_lat_lng,on="Postal Code")
result.head(100)
```

Filter on Boroughs containing Toronto

```python
Toronto_data = result[result.Borough.str.contains("Toronto")]
```

```python
Toronto_data 
```

Let's get the geographical coordinates of Toronto

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

create map of Toronoto using latitude and longitude values

```python
# create map of Toronoto using latitude and longitude values
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(Toronto_data['Latitude'], Toronto_data['Longitude'], Toronto_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
